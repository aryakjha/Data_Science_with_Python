```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-10,10,1000)
y = np.where(x>0,x,0)
```

```
plt.plot(x,y)
plt.grid(True)
plt.title('Activation Relu')
```

```
sigmoid = 1/(1+np.exp(-x))
```

```
plt.plot(x,sigmoid)
plt.grid(True)
plt.title('Activation Sigmoid')
```

```
tanh = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```
plt.plot(x,tanh)
plt.grid(True)
plt.title('Activation tanh')
```

```
xnew = np.random.random((1000,10))
ynew = np.random.random((1000,1))
ynew = np.where(y>0.5,1,0)
```

```
plt.plot(xnew,ynew)
plt.grid(True)
```

```
newmodel = Sequential()
```

```
newmodel.add(Dense(100,input_dim = 10))
newmodel.add(Dense(10))
newmodel.add(Dense(1))
newmodel.add(Activation('sigmoid'))
```

```
newmodel.summary()
```

```
newmodel.compile(loss = 'binary_crossentropy',optimizer = 'adam', metrics = ['accuracy'])
```

```
newmodel.fit(xnew,ynew,epochs = 100)
```

```
pred = newmodel.predict(xnew)
pred
```

```
classes = np.where(pred!=0.47275448,1,0)
print(classes)
```

```
print('accuracy : ', accuracy_score(ynew,classes))
```
