There are 6 types of plots
1. Bar graph

```python
from matplotlib import pyplot as plt
from matplotlib import style

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]

plt.bar(x1,y1,color='g',label = 'bar one')
plt.bar(x2,y2,color='r',label = 'bar two')

plt.title('sample')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend(loc= 'lower right')

plt.grid(True,color='k')
plt.show()
```

2. Histogram

```python
import matplotlib.pyplot as plt

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]

plt.hist(x1, y1,label = 'first one', histtype = 'bar', rwidth = 0.8)
plt.hist(x2, y2,label = 'second one', histtype = 'bar', rwidth = 0.8)
plt.xlabel('X')
plt.ylabel('Y')

plt.title('Histogram')

plt.legend(loc = 'upper right')

plt.grid(True,color='k')

plt.show()
```

3. Scatter plot

```python
import matplotlib.pyplot as plt

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]


plt.scatter(x1,y1,label= 'skt1', color = 'b',s=25,marker = 'o')
plt.scatter(x2,y2,label= 'skt2', color = 'r',s=25,marker = 'o')
plt.xlabel('X')
plt.ylabel('Y')

plt.title('scatter plot')
plt.legend(loc='lower right')
plt.grid(True,color = 'k')
plt.show()
```

4. Stack Plot

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]     #here in this the variables ie the list must be of same length.     
y1 = [2,3,4,3,2]
x2 = [5,10,39,4,3]   
y2 = [6,10,20,4,5]

plt.plot([],[],color='m',label='x2', linewidth = 5)
plt.plot([],[],color='c',label='y1', linewidth = 5)
plt.plot([],[],color='r',label='y2', linewidth = 5)

plt.stackplot(x1,x2,y1,y2,colors = ['m','c','r'])

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Area plot')
plt.legend()
plt.grid(True,color = 'k')
plt.show()



```

Working with multiple Plots

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(221)     #221 means - 2(graphs horizontally), 2(total 2 graphs), 1(this is the first graph)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(222)
plt.plot(t2, f(t2), np.cos(2*np.pi*t2))
plt.show()
```

1. Series

```python
import pandas as pd
import numpy as np

s = pd.Series([1,2,3,4,5,np.nan,6,7,8,9])  #np.nan is for null value
s
```

```python
#we can access the values and index of the series by values and index attributes
import pandas as pd
import numpy as np

s = pd.Series([1,2,3,4,5,np.nan,6,7,8,9])
print(s.values)
print(s.index)
s[3]
s[2:6] # 6 not included
```

```python
#the index need not be an integer, but can consist of values of any desired type. 
#For example, if we wish, we can use strings as an index

import pandas as pd
import numpy as np

s = pd.Series([1.5,2,3,4], index=['a','b','c','d'])
print(s)
print(s[3])   #And the item access works as expected
print(s['d'])
```

```python
#We can even use non-contiguous or non-sequential indices:
import pandas as pd
import numpy as np

s = pd.Series([1.5,2,3,4], index=[3,4,58,10])
s
print(s[58])
```

```python
#Series as specialized dictionary
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
print(population)
print(population['Florida'])
print(population['Texas':'Florida'])
```

2. DataFrame - DataFrame is an analog of a two-dimensional array with both flexible row indices and flexible column names.


```python
import pandas as pd
import numpy as np

area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)
print(area)
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
print(population)

states = pd.DataFrame({'population':population,'area':area})
states               #"column name": object name
```

```python
print(states.index)
print(states.columns)
# Thus the DataFrame can be thought of as a generalization of a 
# two-dimensional NumPy array, where both the rows and columns have a 
# generalized index for accessing the data.
```
